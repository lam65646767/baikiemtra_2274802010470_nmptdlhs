15. Theo thang đo kết quả tính BMI của châu Á:
Chỉ số BMI dưới 18,5 là thiếu cân, 18.5-22,9 là bình thường, 23-24,9 là thừa 
cân, từ 25 trở lên là béo phì.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dữ liệu
data = {
    "Name": ["Mr. Harris", "Mrs. Bradley", "Miss. Laina", "Mrs. Heath", "Mr. William", "Mr. James", "Mr. Timothy", 
             "Mrs. Oscar", "Mrs. Nicholas", "Miss. Marguerite", "Mr. William", "Miss. Hulda", "Mr. Lawrence", "Mr. Wilam"],
    "Height": [1.80, 1.61, 1.78, 1.74, 1.55, 1.90, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    "Weight": [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    "Age": [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
}

# Tạo DataFrame & xử lý dữ liệu
df = pd.DataFrame(data).drop_duplicates(subset="Name", keep='first')
df.loc[df["Age"] > 40, "Age"] = df["Age"].mean()
df["cs_BMI"] = df["Weight"] / (df["Height"] ** 2)

# Phân loại BMI theo chuẩn châu Á
df["Phan_Loai"] = df["cs_BMI"].apply(lambda bmi: '0' if bmi < 18.5 else '1' if bmi <= 22.9 else '2' if bmi <= 24.9 else '3')

# Xuất dữ liệu
df.to_csv("data.csv", index=False)
df.to_excel("data.xlsx", index=False)

# Hiển thị DataFrame
print(df)
```

17. Cho biết:
+ ‘Mr‘tương ứng với ‘male’,
+ ’Mrs’ và ‘Miss’ tương ứng với ‘female’. 
Tạo cột với tên <GT>, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong 
cột <Name> còn các từ đó.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dữ liệu
data = {
    "Name": ["Mr. Harris", "Mrs. Bradley", "Miss. Laina", "Mrs. Heath", "Mr. William", "Mr. James", "Mr. Timothy", 
             "Mrs. Oscar", "Mrs. Nicholas", "Miss. Marguerite", "Mr. William", "Miss. Hulda", "Mr. Lawrence", "Mr. Wilam"],
    "Height": [1.80, 1.61, 1.78, 1.74, 1.55, 1.90, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    "Weight": [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    "Age": [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
}
# Tạo DataFrame
df = pd.DataFrame(data)
# Loại bỏ các hàng trùng lặp dựa trên cột Name, giữ lại hàng đầu tiên
df = df.drop_duplicates(subset=["Name"], keep='first')
# Thay thế các giá trị của cột Age lớn hơn 40 bằng giá trị trung bình của cột đó
df.loc[df["Age"] > 40, "Age"] = df["Age"].mean()
# Thêm cột chỉ số BMI
df["cs_BMI"] = df["Weight"] / (df["Height"] ** 2)
# Phân loại BMI theo tiêu chuẩn châu Á
def classify_bmi(bmi):
    if bmi < 18.5:
        return '0'  # Thiếu cân
    elif 18.5 <= bmi <= 22.9:
        return '1'  # Bình thường
    elif 23 <= bmi <= 24.9:
        return '2'  # Thừa cân
    else:
        return '3'  # Béo phì
df["Phan_Loai"] = df["cs_BMI"].apply(classify_bmi)
# Chuẩn hóa giá trị của cột cs_BMI
df["cs_BMI"] = (df["cs_BMI"] - df["cs_BMI"].min()) / (df["cs_BMI"].max() - df["cs_BMI"].min())
# Tạo cột GT (giới tính)
df["GT"] = df["Name"].apply(lambda x: "male" if "Mr." in x else "female")
# Xóa các tiền tố Mr, Mrs, Miss khỏi cột Name
df["Name"] = df["Name"].str.replace(r'^(Mr\.|Mrs\.|Miss\.)\s*', '', regex=True)
# Hiển thị DataFrame
print(df)
```
